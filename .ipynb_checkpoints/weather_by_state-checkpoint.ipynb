```python
# Dependencies
import csv
import requests as req
import pandas as pd
import json
```

```python
# Save config information.

url = "http://api.wunderground.com/api/cb6b6b20cba56329/almanac/q/"


print(url)

```

```python
weather_data = []
cities = ["NY/New York","VA/Richmond","GA/Atlanta","CA/Los Angeles", "MD/Baltimore", "Birmingham", "AK/Anchorage", "Phoenix", "Little Rock"
         "Denver", "Bridgeport", "Wilmington", "Jacksonville", "Honolulu", "Boise", "Chicago", "Indianapolis", "Des Moines",
          "Wichita", "Louisville", "New Orleans", "Portland", "Boston", "Detroit", "Minneapolis", "Jackson", "Kansas City",
         "Billings", "Omaha", "Las Vegas", "Manchester", "Newark", "Albuquerque", "Charlotte", "Fargo", "Columbus", "Oklahoma City",
         "Portland", "Philadelphia", "Providence", "Charleston", "Sioux Falls", "Nashville", "Houston", "Salt Lake City",
         "Burlington", "Seattle", "Charleston", "Milwaukee", "Cheyenne"]

# Loop through the list of cities and perform a request for data on each
for city in cities:
    response = req.get(url + city + ".json").json()
    weather_data.append(response)

weather_data
```

```python
# Extract interesting data from responses
city_data = [data.get("coord").get("lat") for data in weather_data]
temp_data = [data.get("main").get("temp") for data in weather_data]

weather_data = {"temp": temp_data, "lat": lat_data}
weather_data = pd.DataFrame(weather_data)
weather_data.head()
```
